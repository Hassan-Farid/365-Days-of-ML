# Recommender Systems:
Recommender Systems are systems designed to provide recommendations to the users of the platform, for which it has been designed, and uses various factors and dimensions to select these recommendations. These dimensions may include previous experiences from the user, the category of items or content that they prefer, the preferences of users with similar interests, and a load of other factors depending on the use case. Since these systems allow to provide a good idea of the preferences of a user on so many different factors (levels of computation that humans can't perform), these are usually used with content and product providers the most, including Live streaming sites, E-commerce sites, Gaming communtity, etc. It wouldn't be wrong to say that recommenders have become a part of these processes as a whole.

## General Architecture of Recommender Systems:
Recommender Systems are used for different use cases and there are different designs out there present for different platforms, but a recommender system model is must to comprise of the following components as part of its general architecture:



### Feature Extraction Phase:
This phase comprises of taking data for the user, the items he likes or dislikes and in some cases the context in which the items were rated by the user and performs extraction of features so as to represent the user's interests in the optimal way possible. The extracted features are then fed to the model's retrieval stage, whose task is to select the top N items that are closest to what the user is interested in from millions of items stored in the candidate database. 

### Candidate Retrieval:
This phase takes the features from the feature extraction phase and makes use of different similarity techniques to determine the closest items to the items that user likes or dislikes and making use of them accordingly. The process returns as a final output a filtered list of N items that meet the user's profile. This stage can take place by making use of multiple candidate generators to get the list.

### Candidate Ranking:
Once the top N candidates have been selected, the items are sorted on the basis of their similarity score with the user profile. Based on the metric being used, we can determine which items in the list are relatively more similar to the user profile compared to others and thus are ranked higher in the list for the recommendations.

### Candidate Re-ranking:
From the top N candidates, we again perform ranking but now based on the set of different criteria rather than user profile. These criteria may include highly rated, user context, time of the year, freshness and many other perspectives that allow to narrow down the list even further and give the shows the optimal ranking for the candidate

## Types of Recommender Systems:
Recommender Systems can be categorized into three major types on the basis of how the recommenders utilize the information of its users to find the best recommendation for them. These are as under:

* Content Oriented Systems
* Collaborative Oriented Systems
* Context Oriented Systems

### Content Oriented Systems:
Content Oriented or Content Filtering Recommendation Systems are the type of recommender systems that make use of the users previous choices and preferences and extracts the features from those choices to make decisions for the recommendations to provide. These features may include genre, actors, central idea, etc. to determine what a user likes or dislikes. Using these features, items with similar features are found and are used for recommendation purposes.

A very basic example of content filtering recommenders would be given that a user A likes movies of genres X and Y, the system would recommend it a movie Z with the same genres.

### Collaborative Oriented Systems:
Collaborative Oriented or Collaborative Filtering Recommendation Systems are the type of recommender systems that make use of the preferences and choices of other users with similar interests as the current user and utilize them to offer recommendations to him. The goal is to find the users that performed similar interactions with the system as the current user and generate future predictions on what the current user may like or dislike based on them.

A very basic example of collaborative filtering recommenders would be given that a user A likes movies X and Y and that another user B also likes X and Y, then the system would recommend user A the movie Z, which was liked by user B.

### Context Oriented Systems:
Context Oriented or Context Filtering Recommendation Systems are the type of recommender systems that make use of context information to assign recommendations for users. It makes use of sessions and timing to find patterns on the required preferences of the user. The features utilized in the scenario may include time of the day, season, fulfilment of queries, etc.

A very basic example of where context filtering recommenders would be helpful is buying items from a store, where if the user A whose requirement was to buy some product X got fulfiled, would change his preferences towards product Y instead, removing any trace for desirement of product X till the next interval of its requirement.
